### There are approximitely 3 web development cohorts for every 1 data science cohort at Codeup. Assuming that Codeup randomly selects an alumni to put on a billboard, what are the odds that the two billboards I drive past both have data science students on them?

```python
# 0 is equal to a web development cohort student ---- 1 is equal to a data science cohort student
outcomes = [0,1]
n_simulations = 1_000_000
n_trials = 2

billboards = np.random.choice(outcomes,  size=(n_simulations,n_trials), p=[0.75, 0.25])
```

```python
sum_of_billboards = billboards.sum(axis = 1)
```

```python
(sum_of_billboards == 2).mean()
```

### If 23 people are in the same room, what are the odds that two of them share a birthday? What if it's 20 people? 40?

```python
same_bday = pd.DataFrame(same_birthday)
```

```python
same_bday
```

```python
same_bday["same_birthday"] = same_bday.apply(lambda row: len(row) != row.nunique(),axis = 1)
```
